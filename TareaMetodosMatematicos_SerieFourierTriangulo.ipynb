```python
%pylab inline
from ipywidgets import StaticInteract, RangeWidget
```

```python
#FUNCIONES

# función que da el n-ésimo termino de la serie de fourier
def sumando(y,n):
    z = (8.0/((pi**2)*n**2))*(sin(n*pi/2.0))*sin(n*pi*y/L)*(h)
    return z
```

```python
N_x= 440   #partición de la cuerda
x_i= 0    # punto inicial de la cuerda
L = 1     # punto final de la cuerda
h = 0.5   # altura a la que se estira la cuerda
x_m = 0.5 # punto donde se encuentra la máxima altura de la cuerda

x = linspace(x_i,L,N_x)
f = zeros(N_x)
```

```python
def grafica_fourier(N):
    fig, ax = plt.subplots(figsize=(12, 8),
                           subplot_kw={'axisbg':'#EEEEEE',
                                       'axisbelow':True})
    ax.grid(color='w', linewidth=2, linestyle='solid')
    
    f = zeros(N_x)
    
    # función que suma hasta el termino 
    for i in range(1,N):
        for j in range(N_x):
            f[j] = f[j] + sumando(x[j],i)   
    
    ax.plot(x, f, color='g', lw=2, label='N='+str(N))
    ax.set_xlim(x_i, L)
    ax.set_ylim(0, h)
    
    legend()
    
    return fig
```

```python
StaticInteract(grafica_fourier,
               N=RangeWidget(1, 20, 1))
```

```python
StaticInteract(grafica_fourier,
               N=RangeWidget(20, 201, 20))
```
