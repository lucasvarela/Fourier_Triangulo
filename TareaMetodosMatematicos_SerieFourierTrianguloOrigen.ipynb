```python
%pylab inline
from ipywidgets import StaticInteract, RangeWidget
```

```python
#FUNCIONES

# función que da el n-ésimo termino de la serie de fourier
def funcion(t,n):
    
    suma = 0
    
    for i in range(0,n):
        suma += cos((2*i+1)*pi*t)/(((2*i+1)**2)*(pi**2))
    
    z = suma
    
    return z
```

```python
ts = linspace(-0.5,0.5,100)
```

```python
def grafica_fourier(N):
    fig, ax = plt.subplots(figsize=(12, 8),
                           subplot_kw={'axisbg':'#EEEEEE',
                                       'axisbelow':True})
    ax.grid(color='w', linewidth=2, linestyle='solid')
    
    f = 0.5*ones(100) - 4*funcion(ts,N)
    
    ax.plot(ts, f, color='g', lw=2, label='N='+str(N))
    ax.set_ylim(0, 0.5)
    
    legend()
    
    return fig
```

```python
StaticInteract(grafica_fourier,
               N=RangeWidget(1, 20, 1))
```

```python
StaticInteract(grafica_fourier,
               N=RangeWidget(20, 201, 20))
```
